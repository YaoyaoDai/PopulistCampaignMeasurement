```python
import pandas as pd
import numpy as np
import csv
import json
import string
import random
import re

np.random.seed(1234)
random.seed(1234)
```

```python
SubSpeech_predicted = pd.read_csv('SubSpeech_predicted.csv')
```

```python
Speech = pd.read_csv('CampaignSpeech.csv')
```

```python
#Speech.head()
```

```python
#SubSpeech_predicted.head()
```

```python
Pop_prop = []
for x in Speech.Speech_index:
    Pop_prop.append(np.mean(SubSpeech_predicted.Pop_class[SubSpeech_predicted.Speech_index == x]))
```

```python
Pop_Speech = Speech.assign(Pop_prop = Pop_prop)
```

```python
#Speech.head()
```

```python
Pop_Speech = Pop_Speech[['Candidate','Year','Month','Day','Speech_index','Title','Pop_prop']]
```

```python
#Pop_Speech.head()
```

```python
Pop_Speech.to_csv("Speech_Level_Pop.csv")
```

```python
Pop_Speech.sort_values(by=['Pop_prop'], ascending=False).head(10)
```

```python
Pop_Speech.Month.unique()
```

```python
Candi_Year = pd.DataFrame(columns = ['Candidate', 'Year', 'Pop_prop', 'Counts'])
```

```python
Candidates = list(Pop_Speech.Candidate.unique())
```

```python
for c in Candidates:
    data = Pop_Speech[Pop_Speech.Candidate == c]
    years = list(data.Year.unique())
    for y in years:
        d = data[data.Year==y]
        counts = len(d.Pop_prop)
        prop = np.mean(d.Pop_prop)
        Candi_Year = Candi_Year.append({'Candidate': c, 'Year': y, 'Pop_prop':prop, 'Counts':counts}, 
                                       ignore_index=True)
```
